```python
# inport libraries
import pandas as pd
import numpy as np
# show displays
pd.options.display.max_rows = 999
pd.options.display.max_columns = 100
# set up DF
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python')
# print shape
print(df.shape)
# print head
df.head()
```

```python
# get the first item from the column term
trm=df['term'][:10].iloc[0]
# print trm
print(trm)
```

```python
##### Begin Working Here #####
# Take off the " months" from the column term
def str_to_float(term_str):
  return float((term_str).strip().strip(' months'))
```

```python
df['term'].apply(str_to_float)
```

```python
# new function that will change "Current" or "Fully Paid" to 1 else 0
def Current_or_fully_paid_to1 (status):
  if status== "Current" or "Fully Paid":
    return 1
  else:
    return 0
```

```python
# creates new column from loan status applys function current or fully paid...
df['loan_status_is_great']=df['loan_status'].apply(Current_or_fully_paid_to1)
# show head of df
df.head()
# df.describe('loan_status_is_great')
```

```python
# return the value as a float
str_to_float(trm)
```

```python
# change the date time to legible format
df['last_pymnt_d']=pd.to_datetime(df['last_pymnt_d'],infer_datetime_format=True)
```

```python
# show the  year of each row
df['last_pymnt_d'].dt.year
```

```python
# Make last_pymnt_d_month and last_pymnt_d_year columns.
df['last_pymnt_d_year']=df['last_pymnt_d'].dt.year
```

```python
# New column from last_pymnt_d and print head
df['last_pymnt_d_month']=df['last_pymnt_d'].dt.month
df.head()
```
